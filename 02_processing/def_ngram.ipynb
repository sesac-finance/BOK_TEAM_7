```python
def n_gram_tokenize(csv_loc):
  train_data = pd.read_csv(csv_loc)
  df_bond = pd.DataFrame({'date','content'})
  mpck = MPCK()

  stop_words = ['NR','VV','VA',
                'VX', 'VCP', 'VCN', 'MM', 'MAJ', 'IC',
                'JKS', 'JKC', 'JKG', 'JKO', 'JKB',
                'JKV', 'JKQ', 'JX', 'JC', 'EP', 'EF',
                'EC', 'ETN', 'ETM', 'XPN', 'XSN',
                'XSV', 'XSA', 'XR', 'SF', 'SE',
                'SSO', 'SSC', 'SC', 'SY', 'SL',
                'SH', 'SN']
  bond_tokens_list = []
  final_bond = []
  ngram_result= {}
  bond_text = train_data['content']


  print(type(bond_text))
  print(len(bond_text))
  for i in range(len(bond_text)):
    bond_token = mpck.tokenize(str(bond_text.loc[i]))
    bond_tokens_list.append(bond_token)
    bond_stopword_result = []
    #print(bond_token)

    for bond in bond_tokens_list[i]:
        #print(bond)
        bond_ = bond.split('/')[1]
        # print(bond_)
        for stop in stop_words:
            if bond_ != stop:
                if stop == stop_words[-1]:
                    bond_stopword_result.append(bond)
                else:
                    pass 
            else:
                break


    final_bond.append((bond_stopword_result))
                # dict = {len(bond_text.loc[i]):fianl_bond[i]}
  for j in range (len(final_bond)):
    ngram_result.update({j:mpck.ngramize(final_bond[j])})

  df_ngram = pd.Series(ngram_result)
  # print(df_ngram)
  print(df_bond)
  result_bond = pd.concat([train_data, df_ngram],axis=1)
  result_bond.rename(columns = {0:'ngram'},inplace=True)

  result_bond.to_csv('../data/bok_ngram11111.csv')

  return print('csv파일 생성 완료!')
```
