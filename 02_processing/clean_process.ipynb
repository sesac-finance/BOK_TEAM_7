## 뉴스 간단한 클렌징

```python
import pandas as pd
import requests
import re
import datetime
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/infomax_data.csv')
df.head()
len(df)
```

```python
dong = ('동영상 뉴스', '')
flash = ('flash 오류를 우회하기 위한 함수 추가', '')
end = ('(끝)', '')

s = df['article']
for r in (dong, flash, end):
    s = s.replace(*r)
type(s)
```

```python
s = s.dropna()
s.shape
```

```python
tmplist = []
for i in range(len(s)):
    tmplist.append(re.sub("[\t\n\r]","",re.sub("[^a-zA-Z가-힣ㄱ-ㅎㅏ-ㅣ\s]", "", s.iloc[i])))
    # if i%3000 == 0:
    #     print('%d : %.2f%% 완료됨.' % (i, 100 * i / len(s)))
len(tmplist)
```

```python
pattern = re.compile('[(com)(cokr)].+')
re.sub(pattern,'',tmplist[2000])
```

```python
pattern = re.compile('[(com)(cokr)].+')
finlist = []
for ele in tmplist:
    finlist.append(re.sub(pattern, '', ele))
len(finlist)
```

```python
finlist[2000]
```

```python
processed = pd.Series(finlist)
processed
```

```python
lastlist = []
for ele in finlist:
    lastlist.append(re.sub('서울연합인포맥스','',ele))
final_processed = pd.Series(lastlist)
final_processed
```

```python
final_processed
```

```python
df['processed'] = final_processed
df.shape
```
